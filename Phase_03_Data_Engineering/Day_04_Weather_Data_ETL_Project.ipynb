```python
# ==========================================================
# WEATHER DATA ETL PROJECT
# ==========================================================

# Import Libraries

import pandas as pd

import requests

import warnings

warnings.filterwarnings('ignore')

print("Libraries imported successfully!")
```

```python
# ==========================================================
# EXTRACT — Fetch Weather Data from API
# ==========================================================

city = "Delhi"

api_key = "e3671843bfecc36b02c35e351a5815ea"

url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"

response = requests.get(url)

print("Status Code:", response.status_code)
```

```python
# ==========================================================
# TRANSFORM — Convert JSON Data to Dictionary
# ==========================================================

weather_data = response.json()

print(weather_data)
```

```python
# ==========================================================
# EXTRACT REQUIRED WEATHER DETAILS
# ==========================================================

data = {
    'City': [weather_data['name']],

    'Temperature': [weather_data['main']['temp']],

    'Humidity': [weather_data['main']['humidity']],

    'Weather': [weather_data['weather'][0]['description']],

    'Wind Speed': [weather_data['wind']['speed']]
}

print(data)
```

```python
# ==========================================================
# LOAD — Convert Dictionary to DataFrame
# ==========================================================

df = pd.DataFrame(data)

print(df)
```

```python
# ==========================================================
# SAVE CLEANED WEATHER DATA
# ==========================================================

df.to_csv('weather_data.csv', index=False)

print("Weather data saved successfully!")
```
