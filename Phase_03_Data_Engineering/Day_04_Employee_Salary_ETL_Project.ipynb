```python
# ==========================================================
# EMPLOYEE SALARY ETL PROJECT
# ==========================================================

import pandas as pd

import warnings

warnings.filterwarnings('ignore')

print("Libraries imported successfully!")
```

```python
# ==========================================================
# LOAD EMPLOYEE DATASET
# ==========================================================

df = pd.read_excel('employee_salary_data.xlsx')

print("Dataset Loaded Successfully!")

print("\nFirst 5 Rows:\n")

print(df.head())
```

```python
# ==========================================================
# DATASET OVERVIEW
# ==========================================================

print("Number of Rows:", df.shape[0])

print("Number of Columns:", df.shape[1])

print("\nColumn Names:\n")

print(df.columns.tolist())
```

```python
# ==========================================================
# CHECK DUPLICATE RECORDS
# ==========================================================

print("Duplicate Rows:")

print(df.duplicated().sum())
```

```python
# ==========================================================
# REMOVE DUPLICATE RECORDS
# ==========================================================

df.drop_duplicates(inplace=True)

print("Duplicates removed successfully!")

print("Updated Shape:", df.shape)
```

```python
# ==========================================================
# CALCULATE YEARLY SALARY
# ==========================================================

df['yearly_salary'] = df['monthly_salary'] * 12

print(df[['employee_name', 'monthly_salary', 'yearly_salary']])
```

```python
# ==========================================================
# SAVE CLEANED DATASET
# ==========================================================

df.to_csv('cleaned_employee_salary_data.csv', index=False)

print("Cleaned employee salary dataset saved successfully!")
```
